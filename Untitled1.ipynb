# Arquitectura propuesta para proyecto ML en AWS

Se construye arquitectura de proyecto que pretende la explotación y análisis de información desde diferentes fuentes como: Twitter, correos electrónicos, archivos almacenados en Bucket de S3, entre otros.

Para ello, se propone el diseño e implementación de API que permita controlar el llamado y flujo de peticiones sobre estos servicios, una vez sean recibidos estos insumos pasarán por una base de datos Amazon DynamoDB dónde será posible su aprovechamiento para el almacenamiento de metadatos y datos propios del flujo entrante. Posteriormente, el flujo de datos pasará por un sistema dual de servicios propios de Amazon Kinesis Data: Streams y Firehose, los cuales permiten la transferencia y carga del flujo de datos los cuales estarán monitoreados por CloudWatch, estos estarán conectados a una lambda la cual permitirá realizar las transformaciones necesarias para ser almacenados satisfactoriamente en una instancia RDS y un bucket de S3.

Luego, los datos cargados en el bucket de S3 podrán ser explotados por Amazon Athena para ser analizados por Quicksight, finalmente un componente Glue permitirá catalogar y transformar estos datos, estos análisis podrán ser explorados mediante la API principal.

Por otro lado, se propone un flujo adicional que permita un análisis más profundo del flujo entrando en un bloque de servicios, en este caso: Amazon Kinesis Analitycs, Amazon Comprenhed y Amazon SageMaker, esto gracias a un núcleo multifuncional que conjugue el análisis y el aprendizaje automático, su metadata será llevada a un bucket S3, mientras que sus resultados e interacción con el actor serán mediante un servicio Redshift que permita su consulta y almacenamiento fiable, escalable y robusto.
